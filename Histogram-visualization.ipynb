# What is a Histogram?

- A histogram is a graphical representation of the distribution of numerical data. 
- It is an estimate of the probability distribution of a continuous variable and is constructed by dividing the range of the data into bins (intervals) and counting the number of data points (frequency) that fall into each bin.

    - X-axis: Represents the bins (intervals) of data.
    - Y-axis: Represents the frequency of data points within each bin.

**Usecases**
- Histograms are used to visualize the shape, spread, and central tendency of the data, helping to understand the underlying distribution (e.g., normal, skewed, or multimodal).

# How does Histogram works?

## how a histogram with 2 bins works is demonstrated below

```python
import matplotlib.pyplot as plt
import numpy as np

# To generate same random value in every run
np.random.seed(42)

# Generate random data
data = np.random.randn(1000)
# print(data)
len_data = len(data)

# we are going to generate a histogram with 2 bins
no_of_bins = 2

# RANGE
max_val = data.max()
min_val = data.min()

# BIN WIDTH or jumping value between the bins
range_val = (max_val - min_val)/no_of_bins

# BIN EDGES

# first bin starting value is min_val
# first bin ending value and second bin starting value is below
first_bin_end_val = min_val + range_val

# second bin ending value is below
second_bin_end_val = first_bin_end_val + range_val

print("Expected bin ranges: ",min_val,first_bin_end_val,second_bin_end_val)

# Plot histogram
freq, bin_ranges, patches = plt.hist(data, bins=no_of_bins, edgecolor='black')

print("Frequency of the bins: ", freq)
print("Actual bin ranges: ", bin_ranges)

print(f"\nBIN 1: [{min_val},{first_bin_end_val})") 
print(f"BIN 2: [{first_bin_end_val},{second_bin_end_val}]")

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Random Data')

# Show plot
plt.show()
```

## how a histogram with 'n' bins works is demonstrated below

### testing with no_of_bins = 4

```python
import matplotlib.pyplot as plt
import numpy as np

# To generate same random value in every run
np.random.seed(42)

# Generate random data
data = np.random.randn(1000)
# print(data)
len_data = len(data)

# we are going to generate a histogram with 2 bins
no_of_bins = 4

# RANGE
max_val = data.max()
min_val = data.min()

# BIN WIDTH or jumping value between the bins
range_val = (max_val - min_val)/no_of_bins

# BIN EDGES
bin_edges = [min_val]
next_val = 0
min_val_dummy = min_val
for i in range(no_of_bins):
    next_val = min_val_dummy + range_val
    bin_edges.append(next_val)
    min_val_dummy = next_val

print("Expected bin ranges: ",bin_edges)

# Plot histogram
freq, bin_ranges, patches = plt.hist(data, bins=no_of_bins, edgecolor='black')

print("Frequency of the bins: ", freq)
print("Actual bin ranges: ", bin_ranges)

# print(f"\nBIN 1: [{min_val},{first_bin_end_val})") 
# print(f"BIN 2: [{first_bin_end_val},{second_bin_end_val}]")

for i in range(len(bin_edges)-2):
    print(f"BIN {i+1}: [{bin_edges[i]},{bin_edges[i+1]})")  # for the starting bins, the ending range is not inclued ')'
print(f"BIN {i+2}: [{bin_edges[i+1]},{bin_edges[i+2]}]")    # for the last bin, the last value is included ']'

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Random Data')

# Show plot
plt.show()
```
